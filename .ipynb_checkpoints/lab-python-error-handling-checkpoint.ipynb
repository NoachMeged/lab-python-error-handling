```python
# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
```

   - If the user enters an invalid product name (e.g., a product name that is not in the inventory), or that doesn't have stock available, display an error message and ask them to re-enter the product name. *Hint: you will need to pass inventory as a parameter*
   - Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.

4. Test your code by running the program and deliberately entering invalid quantities and product names. Make sure the error handling mechanism works as expected.


````python
#4. Test your code by running the program and deliberately entering invalid quantities and product names. Make sure the error handling mechanism works as expected.
products = ["apple", "banana", "orange"]

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid integer quantity.")
    
    return inventory

#2. Modify the calculate_total_price function to include error handling.

def calculate_total_price(products, inventory):
    total_price = 0.0
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                price = float(input(f"Enter the price for {product}: "))
                
                if price >= 0:
                    
                    total_price += price * inventory[product]
                    valid_input = True
                else:
                    print("Price cannot be negative. Please enter a valid price.")
            except ValueError:
                print("Invalid input. Please enter a valid numeric price.")
    
    return total_price

#3.Modify the get_customer_orders function to include error handling.


def get_customer_orders(inventory):
    customer_orders = []
    
    # Display available products and their quantities
    print("Available products and their quantities:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

    valid_order_number = False

    # Prompt for the number of orders
    while not valid_order_number:
        try:
            num_orders = int(input("Enter the number of different products to order: "))
            if num_orders >= 0:
                valid_order_number = True
            else:
                print("The number of orders cannot be negative. Please enter a valid number.")
        except ValueError:
            print("Invalid input. Please enter a valid integer.")

    # Prompt for each product order
    for _ in range(num_orders):
        valid_product = False
        while not valid_product:
            product_name = input("Enter the product name you want to order: ").strip()
            
            # Check if product is in inventory and has stock available
            if product_name in inventory:
                if inventory[product_name] > 0:
                    valid_product = True
                    customer_orders.append(product_name)  # Add product to orders
                    inventory[product_name] -= 1  # Decrement the inventory
                else:
                    print(f"Sorry, {product_name} is out of stock. Please choose a different product.")
            else:
                print("Invalid product name. Please enter a valid product from the inventory.")

    return customer_orders


def main():
    print("Welcome to the inventory management system!")

    # Step 1: Initialize the inventory
    print("\n--- Initializing Inventory ---")
    inventory = initialize_inventory(products)

    # Step 2: Calculate total price of products
    print("\n--- Calculating Total Price ---")
    total_price = calculate_total_price(products, inventory)
    print(f"Total price of all products: ${total_price:.2f}")

    # Step 3: Get customer orders
    print("\n--- Customer Orders ---")
    orders = get_customer_orders(inventory)
    print(f"Customer orders: {orders}")
    print(f"Remaining inventory: {inventory}")

# Run the program
main()
````
